```python
from textblob import TextBlob
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment
    return sentiment.polarity, sentiment.subjectivity
```

```python
df['sentiment'] = df['review_text'].apply(get_sentiment)
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import squarify

# Create a sample dataset for demonstration purposes
users = pd.DataFrame({
    "users_uuid": ["user1", "user2", "user3", "user4", "user5", "user6"],
    "display_name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "email_address": ["alice@example.com", "bob@example.com", "charlie@example.com", "david@example.com", "eve@example.com", "frank@example.com"],
    "number_of_photos_uploaded": [100, 80, 60, 40, 20, 10]
})

circles = pd.DataFrame({
    "circles_uuid": ["circle1", "circle2", "circle3"],
    "title": ["Foodies", "Travel", "Nature"],
    "number_of_photos_within_circle": [150, 100, 50]
})

circles_membership = pd.DataFrame({
    "user_uuid": ["user1", "user1", "user2", "user2", "user3", "user3", "user3", "user4", "user5"],
    "circle_uuid": ["circle1", "circle2", "circle1", "circle3", "circle1", "circle2", "circle3", "circle2", "circle3"],
    "number_of_photos_uploaded_by_user_for_circle": [50, 20, 30, 10, 25, 15, 20, 5, 10]
})

# Create a directed graph using the NetworkX package
G = nx.DiGraph()

# Add edges to the graph for each circle membership record
for _, row in circles_membership.iterrows():
    G.add_edge(row["user_uuid"], row["circle_uuid"], weight=row["number_of_photos_uploaded_by_user_for_circle"])

# Calculate the eigenvector centrality of each node
centrality = nx.eigenvector_centrality(G, max_iter=500)

# Draw the network graph
pos = nx.spring_layout(G, seed=42)
node_sizes = [centrality[n] * 1000 for n in G.nodes()]
node_colors = [centrality[n] for n in G.nodes()]
edge_widths = [G[u][v]["weight"] / 10 for u, v in G.edges()]
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.Reds)
nx.draw_networkx_edges(G, pos, width=edge_widths, edge_color="gray")
plt.axis("off")
plt.show()

# Plot the bar chart
plt.bar(users["users_uuid"], users["number_of_photos_uploaded"])
plt.xlabel("User UUID")
plt.ylabel("Number of photos uploaded")
plt.title("Top 10 most active users")
plt.xticks(rotation=90)
plt.show()

# Pivot the table to create a matrix of number of photos uploaded by user and circle
pivot_df = circles_membership.pivot_table(index="user_uuid", columns="circle_uuid", values="number_of_photos_uploaded_by_user_for_circle", fill_value=0)

# Plot the heatmap
sns.heatmap(pivot_df, cmap="Blues")
plt.xlabel("Circle UUID")
plt.ylabel("User UUID")
plt.title("Number of photos uploaded by user and circle")
plt.show()

# Aggregate the number of photos uploaded by user and circle
aggregated_df = circles_membership.groupby("circle_uuid")["number_of_photos_uploaded_by_user_for_circle"].sum()

# Plot the treemap
squarify.plot(sizes=aggregated_df.values, label=aggregated_df.index, color=plt.cm.Set3(range(len(aggregated_df))))
plt.axis("off")
plt.title("Distribution of photos uploaded by circle")
plt.show()

```

```python
import streamlit as st
import pandas as pd
import altair as alt

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 11, 12, 13]
})

chart = alt.Chart(df).mark_line().encode(
    x='x',
    y='y'
).properties(
    width=500,
    height=300
)

st.altair_chart(chart, use_container_width=True)
```
